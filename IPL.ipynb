```python
select = df.toss_decision.value_counts()
labels = (np.array(select.index))
sizes = (np.array((select/select.sum())*100))
plt.pie(sizes,labels = labels,autopct = '%2.1f%%',startangle=90)
plt.title("Toss decision percentage",fontweight="bold",fontsize=15)
plt.show()
```

# Is Match_winner same as Toss_winner..

```python
not_same = df[df["toss_winner"] != df["winner"]]
same = df[df["toss_winner"] == df["winner"]]
per_not = round(len(not_same)/len(df)*100)
per_same = round(len(same)/len(df)*100)
lst = [per_not,per_same]
label = ["notsame","same"]

fig = plt.figure(figsize =(6,7))
plt.pie(lst,autopct = "%2.1f%%",labels = label,shadow = True)
plt.legend()
plt.show()
```
